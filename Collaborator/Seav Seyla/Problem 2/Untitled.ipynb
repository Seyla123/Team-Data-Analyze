```python
import pandas as pd
import numpy as np

# Generate synthetic currency data
np.random.seed(0)  # for reproducibility
years = np.random.choice(range(2010, 2023), size=100)
denominations = np.random.choice(['$1', '$5', '$10', '$20', '$50', '$100'], size=100)
production_numbers = np.random.randint(100000, 1000000, size=100)
circulation_figures = np.random.randint(50000, 900000, size=100)

currency_data = {
    'Denomination': denominations,
    'Year': years,
    'Production_Numbers': production_numbers,
    'Circulation_Figures': circulation_figures
}

# Create DataFrame
df = pd.DataFrame(currency_data)

# Save DataFrame to CSV
df.to_csv('currency_data_100.csv', index=False)

print("CSV file 'currency_data_100.csv' with 100 rows has been saved.")
```

```python
import pandas as pd
import numpy as np

# Generate synthetic currency data
np.random.seed(0)  # for reproducibility
years = np.random.choice(range(2010, 2023), size=1000)
denominations = np.random.choice(['$1', '$5', '$10', '$20', '$50', '$100'], size=1000)
production_numbers = np.random.randint(100000, 1000000, size=1000)
circulation_figures = np.random.randint(50000, 900000, size=1000)

currency_data = {
    'Denomination': denominations,
    'Year': years,
    'Production_Numbers': production_numbers,
    'Circulation_Figures': circulation_figures
}

# Create DataFrame
df = pd.DataFrame(currency_data)

# Save DataFrame to CSV
df.to_csv('currency_data_1000.csv', index=False)

print("CSV file 'currency_data_1000.csv' with 1000 rows has been saved.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic sales data
df = pd.read_csv('./currency_data_1000.csv')

# Explore the dataset
print(df.head())
print(df.info())
print(df.describe())
print(df['Product Category'].value_counts())

# Data cleaning and preparation (if necessary)
# No cleaning needed for this synthetic dataset

# Data visualization
# Distribution of Sales Amount
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales Amount'], bins=30, kde=True)
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Count')
plt.show()

# Sales trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Sales Amount', data=df, estimator=sum, ci=None)
plt.title('Total Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Relationship between Customer Age and Sales Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customer Age', y='Sales Amount', data=df, hue='Customer Gender')
plt.title('Relationship between Customer Age and Sales Amount')
plt.xlabel('Customer Age')
plt.ylabel('Sales Amount')
plt.legend(title='Customer Gender')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./currency_data_1000.csv')

# Explore the dataset
print(df.head())
print(df.info())
print(df.describe())

# Data visualization
# Trends in Production Numbers over Years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Production_Numbers', hue='Denomination', data=df)
plt.title('Trends in Currency Production Numbers')
plt.xlabel('Year')
plt.ylabel('Production Numbers')
plt.legend(title='Denomination')
plt.show()

# Distribution of Currency Denominations
plt.figure(figsize=(8, 6))
sns.barplot(x='Denomination', y='Production_Numbers', data=df, estimator=sum)
plt.title('Distribution of Currency Denominations by Production Numbers')
plt.xlabel('Denomination')
plt.ylabel('Total Production Numbers')
plt.show()

# Correlation between Production Numbers and Circulation Figures
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Production_Numbers', y='Circulation_Figures', data=df, hue='Denomination')
plt.title('Correlation between Production Numbers and Circulation Figures')
plt.xlabel('Production Numbers')
plt.ylabel('Circulation Figures')
plt.legend(title='Denomination')
plt.show()
```
