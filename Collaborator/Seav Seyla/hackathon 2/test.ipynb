```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
quarters = pd.date_range(start='2020Q1', periods=8, freq='Q')
actual_cpi_inflation = [0.5, 0.7, 0.6, 0.8, 0.9, 0.85, 0.75, None]  # Last value is the prediction
predicted_cpi_inflation = [None, None, None, None, None, None, None, 0.80]  # Only the prediction

plt.figure(figsize=(10, 6))
plt.plot(quarters, actual_cpi_inflation, marker='o', linestyle='-', color='b', label='Actual')
plt.plot(quarters, predicted_cpi_inflation, marker='x', linestyle='--', color='r', label='Predicted')
plt.xlabel('Quarter')
plt.ylabel('CPI Inflation Rate')
plt.title('Actual vs. Predicted CPI Inflation Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
# Historical data ending on 2024-01-01
dates = pd.date_range(start='2024-01-01', periods=5, freq='QS')  # Quarterly data ending at 2024-01-01
cpi_inflation = [ 0.95, 0.85, 0.9, 0.75, 0.8]


# Create a DataFrame
df = pd.DataFrame({'CPI_Inflation': cpi_inflation}, index=dates)
print(df)
# Next predicted date and value
next_date = dates[-1] + pd.DateOffset(months=3)  # Adding 3 months to get the next quarter
predicted_value = 0.80

print('next : ', next_date)

# Add the new date and predicted value to the DataFrame
df.loc[next_date] = predicted_value

# Display the updated DataFrame
print(df)

# Plot the data including the predicted value
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['CPI_Inflation'], marker='o', linestyle='-', color='b', label='CPI Inflation')
plt.axvline(x=next_date, color='r', linestyle='--', label='Prediction Start')
plt.xlabel('Date')
plt.ylabel('CPI Inflation Rate')
plt.title('CPI Inflation Rate with Predicted Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Historical inflation rates
inflation_rates = [2.1, 1.8, 2.5, 3.0, 2.2]
alpha = 0.3  # Smoothing factor

# Initialize first forecast as the first inflation rate
forecast = inflation_rates[0]

# Apply exponential smoothing
for rate in inflation_rates[1:]:
    forecast = alpha * rate + (1 - alpha) * forecast

print(f"Exponential Smoothing Forecast: {forecast:.2f}")
```
