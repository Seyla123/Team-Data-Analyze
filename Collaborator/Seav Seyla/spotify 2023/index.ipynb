```python
#Top 10 Artists Have Most Streamed Songs in 2023
import seaborn as sns
count_arists = df['artist(s)_name'].value_counts()
top_artists = count_arists.head(10)

plt.figure(figsize=(10,5))
plt.bar(x=top_artists.index, height=top_artists.values)

plt.title("Top 10 Artists Have Most Streamed Songs in 2023")
plt.ylabel("Number of Hits")
plt.xlabel("Artist")

plt.xticks(rotation=45)
plt.tight_layout
plt.show()
```

```python
#Relationship Between The Number Of Artists In A Song and Streams In 2023
artist_count_stream = df.set_index('artist_count')
artist_count_stream = artist_count_stream['streams']

plt.figure(figsize=(20,10))
sns.barplot(x=artist_count_stream.index , y = artist_count_stream.values)
plt.title('Relationship Between The Number Of Atrists In A Song and Streams In 2023')
plt.xlabel('Number of Artists')
plt.ylabel('Streams')

plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.regplot(x = 'streams', y = 'in_spotify_playlists', data = df, scatter = True, fit_reg = True)
plt.title('Relationship Between Most Streamed Songs And Spotify Playlists In 2023')
plt.xlabel('Streams')
plt.ylabel('Number of Spotify Playlists')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
